Basic Tensor Arithmetic
 (The Hadamard Product)




```python
import numpy as np
import torch
import tensorflow as tf
```

```python
#1.) buat matrix X dengan ukuran 2x3 menggunakan numpy
x = np.array([[21,31,41],[51,61,71]])
x
```

```python
#buat matrix X_pt dengan ukuran 2x3 menggunakan pytorch
x_pt = torch.tensor([[10,20,30],[40,50,60]])
x_pt
```

```python
#buat matrix X_tf dengan ukuran 2x3 menggunakan tensorflow
x_tf = tf.Variable([[2,4,6],[8,10,12]])
x_tf
```

```python
#lakukan perkalian matrix X, X_pt dan X_tf dengan suatu skalar
x
x_pt_np = x_pt.numpy()
x_tf
x * x_pt_np * x_tf
```

```python
#lakukan penjumlahan matrix X, X_pt dan X_tf dengan suatu skalar
x + x_pt_np + x_tf
```

```python
#lakukan pengurangan matrix X, X_pt dan X_tf dengan suatu skalar
x - x_pt_np - x_tf
```

```python
#2.) buat matrix Y dengan ukuran 2x3 menggunakan numpy
y = np.array([[9,8,7],[6,5,4]])
y
```

```python
#buat matrix Y_pt dengan ukuran 2x3 menggunakan pytorch
y_pt = torch.tensor([[12,21,23],[34,43,45]])
y_pt
```

```python
#buat matrix Y_tf dengan ukuran 2x3 menggunakan tensorflow
y_tf = tf.Variable([[15,25,35],[42,52,62]])
y_tf
```

```python
#kalikan matrix X dengan Y
x * y
```

```python
#kalikan matrix X_pt dengan Y_pt
x_pt * y_pt
```

```python
#kalikan matrix X_tf dengan Y_tf
x_tf * y_tf
```
