Visualizing Linear Systems

```python
import numpy as np
import matplotlib.pyplot as plt
```

y = 2x+1 , 3x-y = 5

$$ y = 2x+1 $$

$$ 3x-y = 5 $$
$$ -y = 5-3x $$
$$ y = 3x-5$$

```python
#1.) buat visualisasi dari persamaan berikut y = 2x+1 dan 3x-y = 5
x = np.linspace(-10, 10, 1000)
```

```python
y1 = (2*x) + 1
```

```python
y2 = (3*x) - 5
```

```python
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')

plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')


ax.set_xlim([-2, 8])
ax.set_ylim([-6, 20])
ax.plot(x, y1, c='green')
ax.plot(x, y2, c='brown')
plt.axvline(x=6, color='purple', linestyle='--')
_ = plt.axhline(y=13, color='purple', linestyle='--')
```
