Tensor Reduction

```python
#1.) buat matrix X dengan ukuran 2x3 menggunakan numpy
import numpy as np
x = np.array([[10,20,30], [40,50,60]])
x
```

```python
#hitung berapa hasil penjumlahan elemen matrix X dengan fungsi sum
x.sum()
```

```python
#buat matrix X_tf dengan ukuran 2x3 menggunakan tensorflow
import tensorflow as tf
x_tf = tf.Variable([[11,12,13],[14,15,16]])
x_tf
```

```python
#2.) hitung hasil penjumlahan seluruh elemen matrix X
x.sum()
```

```python
#hitung hasil penjumlahan seluruh elemen matrix X_pt
import torch
x_pt = torch.tensor([[52,10,25],[90,60,42.]])
x_pt.sum()
```

```python
#hitung hasil penjumlahan seluruh elemen matrix X_tf
tf.reduce_sum(x_tf)
```

```python
#3.) hitung hasil penjumlahan matrix X berdasarkan row nya
x.sum(axis=1)
```

```python
#hitung hasil penjumlahan matrix X_pt berdasarkan row nya
x_pt.sum(axis=1)
```

```python
#hitung hasil penjumlahan matrix X_tf berdasarkan row nya
tf.reduce_sum(x_tf, axis = 1)
```

```python
#4.) hitung hasil penjumlahan matrix X berdasarkan kolom nya
x.sum(axis=0)
```

```python
#hitung hasil penjumlahan matrix X_pt berdasarkan kolom nya
x_pt.sum(axis=0)
```

```python
#hitung hasil penjumlahan matrix X_tf berdasarkan kolom nya
tf.reduce_sum(x_tf, axis = 0)
```

```python
#5.) hitung nilai minimal dari matrix X, X_pt dan X_tf
print(x.min())
print(x_pt.min())
print(tf.reduce_min(x_tf))
```

```python
#hitung nilai maksimal dari matrix X, X_pt dan X_tf
print(x.max())
print(x_pt.max())
print(tf.reduce_max(x_tf))
```

```python
#hitung nilai rata-rata dari matrix X, X_pt dan X_tf
print(x.mean())
print(x_pt.mean())
print(tf.reduce_mean(x_tf))
```

```python
#6.) hitung hasil perkalian seluruh elemen matrix X
x.prod()
```

```python
#hitung hasil perkalian seluruh elemen matrix X_pt
x_pt.prod()
```

```python
#hitung hasil perkalian seluruh elemen matrix X_tf
tf.math.reduce_prod(x_tf)
```
