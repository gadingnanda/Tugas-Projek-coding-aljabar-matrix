The Dot Product

```python
import numpy as np
import torch
import tensorflow as tf
```

```python
#1.) buat vector x dan y dengan ukuran 1x3 menggunakan numpy
x = np.array([12,5,7])
x
```

```python
#buat vector x_pt dan y_pt dengan ukuran 1x3 menggunakan pytorch
x_pt = torch.tensor([6,9,14])
x_pt
```

```python
#buat vector x_tf dan y_tf dengan ukuran 1x3 menggunakan tensorflow
x_tf = tf.Variable([22,4,10])
x_tf
```

```python
#2.) hitung perkalian dot product antara vector x dan vector y
y = np.array([3,7,9])
np.dot(x,y)
```

```python
#hitung perkalian dot product antara vector x_pt dan vector y_pt
y_pt = torch.tensor([10,2,15])
torch.dot(x_pt,y_pt)
```

```python
#hitung perkalian dot product antara vector x_tf dan vector y_tf
y_tf = tf.Variable([2,6,10])
tf.reduce_sum(tf.multiply(x_tf,y_tf))
```
