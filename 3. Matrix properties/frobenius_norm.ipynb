Frobenius Norm

```python
import numpy as np
import torch
import tensorflow as tf
```

```python
#1.) menggunakan numpy, pytoch dan tensorflow buatlah matrix ukuran 3x3 lalu hitung nilai frobenius norm
#numpy
x = np.array([[2,4,6],[3,7,9],[10,12,15]])
x_pt = torch.tensor([[1,3,5],[10,21,12],[4,9,7.]])
x_tf = tf.Variable([[10,20,30],[12,23,4],[12,22,26.]])

print(np.linalg.norm(x))
print(torch.norm(x_pt))
print(tf.norm(x_tf))
```
