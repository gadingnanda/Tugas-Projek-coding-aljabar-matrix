```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#buat matrix X ukuran 2x2 menggunakan numpy
X = np.array([[2, 6], [-3, -7]])
X = np.matrix(X)
X

```

```python
#buat invers dari matrix X dan beri nama Xinv menggunakan numpy
Xinv = np.linalg.inv(X)
Xinv
```

```python
#kalikan X dengan Xinv menggunakan numpy
np.dot(X, Xinv)
```

```python
#selesaikan persamaan berikut menggunakan numpy, tensorflow dan pytorch

import numpy as np

A = np.array([[4, 2],
              [-5, -3]])
B = np.array([4, -7])

# Mencari solusi x
x = np.linalg.solve(A, B)

print(f"NumPy Solution: b = {x[0]:.2f}, c = {x[1]:.2f}")

import tensorflow as tf

A = tf.constant([[4.0, 2.0],
                 [-5.0, -3.0]])
B = tf.constant([[4.0],
                 [-7.0]])

# Mencari solusi x
x = tf.linalg.solve(A, B)

print(f"TensorFlow Solution: b = {x[0][0]:.2f}, c = {x[1][0]:.2f}")

import torch

A = torch.tensor([[4.0, 2.0],
                  [-5.0, -3.0]])
B = torch.tensor([4.0, -7.0])

# Mencari solusi x
x = torch.linalg.solve(A, B)

print(f"PyTorch Solution: b = {x[0]:.2f}, c = {x[1]:.2f}")
```
