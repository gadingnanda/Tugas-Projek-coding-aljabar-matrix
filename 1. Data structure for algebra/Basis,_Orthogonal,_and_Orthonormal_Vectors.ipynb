Basis, Orthogonal, and Orthonormal Vectors

```python
import numpy as np
```

```python
#buat vector i dengan elemen (1,0) dan vector j dengan elemen (0,1)
i = np.array([1,0])
j = np.array([0,1])
print(i)
print(j)
```

```python
#buktikan bahwa vector i dan vector j itu orthogonal
np.dot(i, j)
```

```python
#2.) Bagaimana dengan v=(1,2) dan w=(2,-1), apakah juga orthogonal?
v = np.array([1,2])
w = np.array([2,-1])
np.dot(v, w)
```
