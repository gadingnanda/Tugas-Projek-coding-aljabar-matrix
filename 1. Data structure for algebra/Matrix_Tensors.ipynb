```python
import numpy as np
```

```python
#1.) buat matrix X berukuran 2x3 menggunakan numpy
x = np.array([[25,15,66], [52,51,66]])
x
```

```python
#tampilkan ukuran dari matrix X
x.shape
```

```python
#tampilkan seluruh elemen dari kolom pertama dari matrix X
x[:,0]
```

```python
#tampilkan seluruh elemen dari baris kedua dari matrix X
x[1,:]
```

```python
#2.) ulangi soal nomor 1 menggunakan pytorch
#ulangi soal nomor 1 menggunakan tensorflow
import torch
import tensorflow as tf
```

```python
#buat matrix X berukuran 2x3 menggunakan pytorch
x_pt = torch.tensor([[25,15,66], [52,51,66]])
x_pt
```

```python
#tampilkan ukuran dari matrix x_pt
x_pt.shape
```

```python
#tampilkan seluruh elemen dari kolom pertama dari matrix x_pt
x_pt[:,0]
```

```python
#tampilkan seluruh elemen dari baris kedua dari matrix x_pt
x_pt[1,:]
```

```python
#buat matrix X berukuran 2x3 menggunakan tensorflow
x_tf = tf.Variable([[25,15,66], [52,51,66]])
x_tf
```

```python
#tampilkan ukuran dari matrix x_tf
x_tf.shape
```

```python
#tampilkan seluruh elemen dari kolom pertama dari matrix x_tf
x_tf[:,0]
```

```python
#tampilkan seluruh elemen dari baris kedua dari matrix x_tf
x_tf[1,:]
```
