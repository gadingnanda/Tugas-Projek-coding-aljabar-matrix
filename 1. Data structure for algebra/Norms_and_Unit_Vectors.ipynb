Norms and Unit Vectors


```python
import numpy as np
```

```python
#1.) buat vector x dengan ukuran 3x1
x = np.array([25,15,66])
x
```

```python
#hitung nilai L2 norm dari vector x
np.linalg.norm(x)
```

```python
#2.) buat vector v dengan elemen (3,4)
v = np.array([3,4])
v
```

```python
#hitung nilai L2 norm dari vector v
np.linalg.norm(v)
```

```python
#tampilkan unit vector dari vector v
norm = np.linalg.norm(v)
unit_v = v / norm
print("Vektor asli:", v)
print("Unit vector:", unit_v)
```

```python
#hitung panjang unit vector dari vector v
print("Panjang vektor:", norm)
print("Panjang unit vector:", np.linalg.norm(unit_v))
```
