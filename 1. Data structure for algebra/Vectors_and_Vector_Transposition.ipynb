```python
#2.) lakukan transpose pada vector x dan simpan pada variabel x_t
x_t = x.T
x_t
```

```python
#tampilkan ukuran dari vector x_t
x_t.shape
```

```python
#lakukan transpose pada vector y dan simpan pada variabel y_t
y = np.array([[25,15,66], [52,51,66]])
y_t = y.T
y_t
```

```python
#tampilkan ukuran dari vector y_t
y_t.shape
```

```python
#3.) lakukan soal nomor 2 menggunakan pytorch dan tensorflow
import torch
import tensorflow as tf
```

```python
#lakukan transpose pada vector x_pt dan simpan pada variabel x_pt_t
x_pt = torch.tensor([25, 15, 66])
x_pt_t = x_pt.T
x_pt_t
```

```python
#tampilkan ukuran dari vector x_pt_t
x_pt_t.shape
```

```python
#lakukan transpose pada vector y_pt dan simpan pada variabel y_pt_t
y_pt = torch.tensor([[25, 15, 66], [52,51,66]])
y_pt_t = y_pt.T
y_pt_t
```

```python
#tampilkan ukuran dari vector y_pt_t
y_pt_t.shape
```

```python
#lakukan transpose pada vector x_tf dan simpan pada variabel x_tf_t
x_tf = tf.Variable([25,15,66])
x_tf_t = tf.transpose(x_tf)
x_tf_t
```

```python
#tampilkan ukuran dari vector x_tf_t
x_tf_t.shape
```

```python
#lakukan transpose pada vector y_tf dan simpan pada variabel y_tf_t
y_tf = tf.Variable([[25,15,66], [52,51,66]])
y_tf_t = tf.transpose(y_tf)
y_tf_t
```

```python
#tampilkan ukuran dari vector y_tf_t
y_tf_t.shape
```
